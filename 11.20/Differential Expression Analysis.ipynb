```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dea_data.csv")
```

```python
df.head()
```

```python
control=[col for col in df.columns if "Control" in col]
treatment=[col for col in df.columns if "Treatment" in col]
```

```python
df['Mean_Control'] = df[control].mean(axis=1)
df['Mean_Treatment'] = df[treatment].mean(axis=1)
```

```python
df['Log2FC']=np.log2(df["Mean_Treatment"]/ df["Mean_Control"])
```

```python
def classify_gene(log2fc):
    if log2fc >= 1:
        return 'Upregulated'
    elif log2fc <= -1:
        return 'Downregulated'
    else:
        return 'Not significant'

df['Status'] = df['Log2FC'].apply(classify_gene)
```

```python
result_df = df[['Gene', 'Mean_Control', 'Mean_Treatment', 'Log2FC', 'Status']]
print(result_df.head())
```
