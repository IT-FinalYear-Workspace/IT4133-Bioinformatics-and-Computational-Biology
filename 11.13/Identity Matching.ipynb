```python
def calculate_identity(seq1, seq2):
    matches = 0
    total = len(seq1)
    for i in range(total):
        if seq1[i] == seq2[i]:
            matches += 1
    identity_percentage = (matches / total) * 100
    return identity_percentage
```

```python
def find_similarity(seq1, seq2):
    matches = 0
    total = min(len(seq1), len(seq2))
    for i in range(total):
        if seq1[i] == seq2[i]:
            matches += 1
    similarity = (matches / total) * 100
    return similarity
```

```python
dna_Seq1='ATGCCGTAGCTAAGGCTT'
mutation_pos=6

sub_mutationDna = dna_Seq1[:mutation_pos] + 'C' + dna_Seq1[mutation_pos+1:]
ins_mutation = dna_Seq1[:mutation_pos] + 'G' + dna_Seq1[mutation_pos:]
dele_mutation = dna_Seq1[:mutation_pos] + dna_Seq1[mutation_pos+1:]  
```

```python
print("Origial DNA Sequence:",dna_Seq1)
print("Substitution Mutation Sequence:",sub_mutationDna)
print("Insertion Mutation Sequence:",ins_mutation)
print("Deletion Mutation Sequence:",dele_mutation)
```

```python
import matplotlib.pyplot as plt

seqs = ["ATGC", "ATGCGG", "GGGCCCATA", "ATATGG"]
gc_vals = []
lengths = []

def gc_content(s):
    gc = 0
    for x in s:
        if x in ["G","C"]:
            gc += 1
    return (gc/len(s))*100

for s in seqs:
    lengths.append(len(s))
    gc_vals.append(gc_content(s))

plt.scatter(lengths, gc_vals)
plt.title("GC Content vs Length")
plt.xlabel("Length")
plt.ylabel("GC%")
plt.show()
```
