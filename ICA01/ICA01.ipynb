```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("dna_samples.csv")
```

```python
df.head()
```

```python
def is_valid_sequence(seq) :
    for i in seq:
        if i not in ["A","T","G","C"]:
            return False
    return True

df["Valid"]= df["DNASequence"].apply(is_valid_sequence)
df.head()
```

```python
clean_df= df[df["Valid"]== True].drop(columns=["Valid"]).reset_index(drop=True)
clean_df.head()
```
