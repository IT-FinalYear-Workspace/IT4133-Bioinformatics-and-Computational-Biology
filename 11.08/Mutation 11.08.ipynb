```python
#Transalte the dna Sequence into Protein
# DNA -> complement -> mRNA -> codons -> amino acids
# DNA -> Complemnt
# A -> T
# T -> A
# G -> C
# C -> G

# Complemnt -> mRNA
# T -> U

# mRNA -> codons
# Seperate to 3 

# codons -> amino acids
# Codon Table

# DNA:  ATG TCC TTT TTA TCG
# mRNA: UAC AGG AAA AAU AGC

```

```python
# dnaSeq=['A','T','G','T','C','C','T','T','T','T','T','A','T','C','G']
dnaSeq='ATGTCCTTTTTATCG'
def convert_Complemnt(dnaSeq):
    complement=""
    for base in dnaSeq:
        if base == 'A':
            complement += "T"
        elif base == 'T':
            complement += "A"
        elif base == 'G':
            complement += "C"
        elif base == 'C':
            complement += "G"
    return complement

def convert_mRNA(dnaSeq):
    mRNA=""
    for base in complement:
        if base == 'A':
            mRNA += "A"
        elif base == 'T':
            mRNA += "U"
        elif base =='G':
            mRNA += "G"
        elif base == 'C':
            mRNA += "C"
    return mRNA
        

complement= convert_Complemnt(dnaSeq)
mRNA= convert_mRNA(complement)
print("Original DNA:       ", dnaSeq)
print("Complementary DNA:  ", complement)
print("mRNA Sequence:      ", mRNA)
        
    
```

```python
def convert_Protein(dnaSeq):
    protein = ""
    codon_list=[]
    for i in range(0, len(mRNA), 3):
        codon = mRNA[i:i+3]
        codon_list.append(codon)

        if codon == 'UAC':
            protein += "Tyr-"
        elif codon == 'AGG':
            protein += "Arg-"
        elif codon == 'AAA':
            protein += "Lys-"
        elif codon == 'AAU':
            protein += "Asn-"
        elif codon == 'AGC':
            protein += "Ser-"
        elif codon == 'AGG':
            protein += "Arg-"


    if protein.endswith("-"):
        protein = protein[:-1]
    print("Codon Sequence: ", '-'.join(codon_list))
    return protein


protein= convert_Protein(mRNA)
print("Protein Sequence:", protein)
```

```python
def point_mutation(dnaSeq, position, new_base):
    if position < 0 or position >= len(dnaSeq):
        return "Error: Position out of range."
    if new_base not in 'ATGC':
        return "Error: Invalid nucleotide."
    
    dna_list = list(dnaSeq)
    dna_list[position] = new_base 
    mutated_seq = ''.join(dna_list)
    return mutated_seq

mutated = point_mutation(dnaSeq, 5, 'A')
print("Mutated DNA Seq: ", mutated)

complement_mutated= convert_Complemnt(mutated)
mRNA_mutated= convert_mRNA(complement_mutated)
mutated_protein= convert_Protein(mRNA_mutated)
print("Complementary Mutated DNA: ", complement_mutated)
print("Mutated mRNA Sequence: ", mRNA_mutated)
print("Mutated Protein Sequence:", mutated_protein)

```
