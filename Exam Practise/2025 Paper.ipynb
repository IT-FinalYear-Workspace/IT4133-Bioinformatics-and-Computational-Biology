```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dna_seq = "AGCTTTCGAATTCGA"
```

```python
palindromes=[]

for i in range(len(dna_seq)):
    for j in range(i+1, len(dna_seq) +1):
        sub = dna_seq[i:j]
        if sub == sub[::-1]:
            if sub not in palindromes:
                palindromes.append(sub)

palindromes
```

```python
def gc_count(seq):
    gc= seq.count('G')+seq.count('C')
    return (gc /len(seq)) * 100

```

```python
for p in palindromes:
    print(p, ": ", round(gc_count(p),2), "%")
```

```python
longest=palindromes[0]
for p in palindromes:
    if len(p) > len(longest):
        longest = p

longest
```

```python
longest=max(palindromes,key=len)
longest
```

```python
df.describe()
```

```python
control_cols= [i for i in df.columns if 'Control' in i]
treatment_cols= [i for i in df.columns if 'Treatment' in i]
df['Control_mean']= df[control_cols].mean(axis=1)
df['Treatment_mean']= df[treatment_cols].mean(axis=1)
print(df['Control_mean'])
print(df['Treatment_mean'])
```

```python
df['Log2Fold']= np.log2(df['Treatment_mean']/ df['Control_mean'])

df['Status']='No Change'
df.loc[df['Log2Fold'] > 1.6, 'Status'] = 'Upregulated'
df.loc[df['Log2Fold'] < 0.67, 'Status'] = 'Downregulated'
df
```

```python
plt.figure()
bars=plt.bar(df['Gene'], df['Log2Fold'])
plt.xlabel("Gene")
plt.xticks(rotation=90)
plt.ylabel("Log2Fold")
plt.title("Fold change of Genes")
plt.legend(bars,['Gene','Treatment'],
           loc='upper left', bbox_to_anchor=(1.05,1))
plt.show()
```

```python
output_df= pd.DataFrame({
    'Gene': df['Gene'],
    'Control_mean': df['Control_mean'],
    'Treatment_mean': df['Treatment_mean'],
    'Fold_Change': df['Log2Fold'],
    'Status': df['Status']
})
output_df
```
