```python
Control_DNA= "ATGGCCATGGCGCCCAGAACCTGAGATCAATCGATGCGGGCTACGGCGACTGGTCGTAG"
Patient_DNA= "ATGGCCATGGCGCCCAGAACCTGAGATCAATCGATGCGAGCTACGGCGACTGGTCGTAG"
introns = [(15, 25)] 
codon_table = {
'AUG': 'M', 'UUU': 'F', 'UUC': 'F', 'UUA': 'L', 'UUG': 'L',
'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S',
'UAU': 'Y', 'UAC': 'Y', 'UAA': '*', 'UAG': '*',
'UGU': 'C', 'UGC': 'C', 'UGA': '*', 'UGG': 'W',
'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
'CAU': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
'AUU': 'I', 'AUC': 'I', 'AUA': 'I', 'ACA': 'T',
'ACC': 'T', 'ACG': 'T', 'ACU': 'T',
'AAU': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K',
'AGU': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R',
'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',
'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
'GAU': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',
'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'
}
```

```python
dnaSeq1='ATGGCTGAACTG'
dnaSeq2='ATGGCCGAATTG'
```

```python
def do_complement(dnaSeq):
    dnaSeq=dnaSeq.upper()
    complementSeq=''
    complent_base={'A':'T','T':'A','G':'C','C':'G'}
    for i in range(len(dnaSeq)):
        complementSeq += complent_base[dnaSeq[i]]
    # complement= ''.join(complent_base[i] for i in dnaSeq)
    return complementSeq

print("Antiparallel Control", do_complement(Control_DNA))
print("Antiparallel Patient", do_complement(Patient_DNA))
```

```python
def do_transcribe(dnaSeq):
    complementSeq= do_complement(dnaSeq)
    return complementSeq.replace('T','U')

# print("Transcribe Control Seq: ", do_transcribe(Control_DNA))
# print("Transcribe Patient Seq: ", do_transcribe(Patient_DNA))
```

```python
def do_splicemRNA(dnaSeq, introns):
    spliced=dnaSeq
    offset=0
    for start , end in introns:
        spliced = spliced[:start - offset] + spliced[end- offset +1:]
        offset += (end - start +1)
    return spliced

# print("Spliced Control: ", do_splicemRNA(Control_DNA,introns))
# print("Spliced Treatment: ", do_splicemRNA(Patient_DNA,introns))
```

```python
def do_translation(mrna):
    protein=''
    for i in range(0,len(mrna),3):
        codon=mrna[i:i+3]
        if len(codon) < 3:
            break

        amino=codon_table.get(codon,'?')

        if amino == '*':
            break

        protein += amino
    return protein
```

```python
def compare_proteins(prot1, prot2):
    differences = []
    length = min(len(prot1), len(prot2))
    for i in range(length):
        if prot1[i] != prot2[i]:
            differences.append((i+1, prot1[i], prot2[i]))
    return differences
```

```python
control_comp = do_complement(Control_DNA)
patient_comp = do_complement(Patient_DNA)
print("Control Complementary DNA:", control_comp)
print("Patient Complementary DNA:", patient_comp)

# Transcription to RNA
control_rna = do_transcribe(Control_DNA)
patient_rna = do_transcribe(Patient_DNA)
print("\nControl RNA:", control_rna)
print("Patient RNA:", patient_rna)

# Splicing RNA
control_mrna = do_splicemRNA(control_rna, introns)
patient_mrna = do_splicemRNA(patient_rna, introns)
print("\nControl mRNA:", control_mrna)
print("Patient mRNA:", patient_mrna)

# Translation to protein
control_protein = do_translation(control_mrna)
patient_protein = do_translation(patient_mrna)
print("\nControl Protein:", control_protein)
print("Patient Protein:", patient_protein)

# Compare proteins
mutations = compare_proteins(control_protein, patient_protein)
print("\nMutations (position, control AA, patient AA):", mutations)
```
