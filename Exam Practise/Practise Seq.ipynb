```python
dna = "AGCTTTCGAATTCGA"
```

```python
#Find all substrings
subtrings=[]

for i in range( len(dna)):
    for j in range(i+1, len(dna)+1):
        subtrings.append(dna[i:j])

subtrings

```

```python
#Find repeated substrings
repeated = []

for i in range(len(dna)):
    for j in range(i+1, len(dna)+1):
        sub = dna[i:j]
        if dna.count(sub) > 1 and sub not in repeated:
            repeated.append(sub)

print(repeated)
```

```python
#Count nucleotide frequency
a = dna.count("A")
t = dna.count("T")
g = dna.count("G")
c = dna.count("C")

print("A:", a, "T:", t, "G:", g, "C:", c)
```

```python
#Remove invalid characters
clean = ""

for ch in dna:
    if ch in "ATGC":
        clean += ch

print(clean)
```

```python
#Check if DNA is palindrome
if dna == dna[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#Palindromes longer than 3
palindromes = []

for i in range(len(dna)):
    for j in range(i+1, len(dna)+1):
        sub = dna[i:j]
        if sub == sub[::-1] and len(sub) > 3:
            if sub not in palindromes:
                palindromes.append(sub)

print(palindromes)
```

```python
#Count palindromes
count = 0

for i in range(len(dna)):
    for j in range(i+1, len(dna)+1):
        sub = dna[i:j]
        if sub == sub[::-1]:
            count += 1

print(count)
```

```python
#Find motif occurrences
motif = "GAATTC"

for i in range(len(dna)-len(motif)+1):
    if dna[i:i+len(motif)] == motif:
        print("Found at", i)
```

```python
#Count motif
count = 0

for i in range(len(dna)-len(motif)+1):
    if dna[i:i+len(motif)] == motif:
        count += 1

print(count)
```

```python
# Find start codon
for i in range(0, len(dna)-2):
    if dna[i:i+3] == "ATG":
        print("Start codon at", i)
```

```python
# Find stop codons
stops = ["TAA", "TAG", "TGA"]

for i in range(len(dna)-2):
    if dna[i:i+3] in stops:
        print("Stop codon at", i)
```

```python
#Sort palindromes by length
palindromes.sort(key=len)
print(palindromes)
```
