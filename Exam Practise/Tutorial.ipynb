```python
reads = [
    "TCATC",
    "TTCAT",
    "TCATC",
    "TGAAA",
    "GAGGA",
    "TTTCA",
    "ATCAA",
    "TTGAT",
    "TTTCC"
]
```

```python
def rev_comp(seq):
    comp={'A':'T','T':'A','G':'C','C':'G'}
    result=''
    for b in reversed(seq):
        result += comp[b]
    return result

def hamming(s1,s2):
    diff=0
    for i in range(len(s1)):
        if s1[i]!= s2[i]:
            diff += 1
        return diff


frequent=[]
for r in reads:
    if reads.count(r) >1 or reads.count(rev_comp(r)) >1:
        frequent.append(r)

for r in reads:
    if r not in frequent:
        for f in frequent:
            if hamming(r,f) ==1:
                print(f"{r} -> {f}")
                break
            elif hamming(r,rev_comp(f)) ==1:
                print(f"{r} -> {rev_comp(f)}")
                break
```
